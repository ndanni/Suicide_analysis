```python
#importing tools for anaylysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#to read in the data
data1 = pd.read_csv('..\data\master.csv')
data2 = pd.read_csv('..\data/annual-number-of-deaths-by-cause.csv')
data3 = pd.read_csv('..\data\share-of-deaths-by-cause-2016.csv')
```

```python
#dropping columns that are not useful in this analysis
to_drop1 = ['country-year', 'HDI for year', ' gdp_for_year ($) ', 'gdp_per_capita ($)', 'generation']
data1.drop(columns=to_drop1, inplace=True)
```

```python
#to view the first five rows of the data so to know if those columns have been dropped.
data1.head()
```

```python
to_drop3 = ['Code']
data3.drop(columns = to_drop3, inplace=True)
```

```python
#to view the minimum and maximum dates of data1
min_year=min(data1.year)
max_year=max(data1.year)
print('Min Year :',min_year)
print('Max Year :',max_year)
```

```python
#to view the minimum and maximum dates of data2
min_year=min(data2.Year)
max_year=max(data2.Year)
print('Min Year :',min_year)
print('Max Year :',max_year)
```

```python
#to view the minimum and maximum dates of data3
min_year=min(data3.Year)
max_year=max(data3.Year)
print('Min Year :',min_year)
print('Max Year :',max_year)
```

```python
#reducing the year in data1 to 1990 in other to have same year values with other data frame.
data1 = data1[data1.year >= 1990]
```

```python
#renaming the column year to Year.
data1.rename(columns={'year':'Year'})
```

```python
#groping the data according to country and sexes so as to grasp an undertsanding of the data
data1.groupby(['country', 'sex', 'year', 'age']).suicides_no.sum()
```

```python
print((data1[data1['sex'] == 'male']))
```

```python
#to view the country with the highest suicide rate
plt.figure(figsize=(50,30))
visual = sns.barplot(x = 'country', y = 'suicides_no', data = data1)
visual.set_title('Total rate of suicides between 1990-2016')
```

```python
#to view suicide rate by age
plt.figure(figsize=(20,15))
visual2 = sns.barplot(x='age', y='suicides_no', data=data1);
visual2.set_title('Suicide rate by age')
```

```python
#to view suicide rate by sex
plt.figure(figsize=(8,10))
figure = sns.barplot(x='sex', y='suicides_no', data=data1);
figure.set_title('Suicide rate by sex')
```

```python
#to view suicde rate by age and sex
plt.figure(figsize=(20,10))
figure2 = sns.barplot(data=data1,x='sex',y='suicides_no',hue='age');
figure2.set_title('suicide rate by sex and age')
```

```python
#to view suicde rate by age and sex
plt.figure(figsize=(20,10))
figure2 = sns.barplot(data=data2,x='country',y='suicides_no',hue='age');
figure2.set_title('suicide rate by sex and age')
```
